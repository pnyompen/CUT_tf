```python
from PIL import Image
from matplotlib import pyplot as plt
import random
from pathlib import Path
import tensorflow as tf
import numpy as np
```

```python
img_path = '/Volumes/T5/Datasets/FFHQ/69918.png'
```

```python

im = Image.open(img_path)
im = np.array(im)
plt.imshow(im)
```

```python
tensor = np.array(im).transpose((2,0,1))
tensor = tf.convert_to_tensor(tensor)
tensor = tf.keras.preprocessing.image.random_rotation(tensor, 25, fill_mode='reflect')
zoom_ratio = random.random() * 1 + 0.75 
tensor = tf.keras.preprocessing.image.random_zoom(tensor, (zoom_ratio, zoom_ratio), fill_mode='reflect')
tensor = np.array(tensor).transpose(1,2,0)

plt.imshow(tensor)
```
