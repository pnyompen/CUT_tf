```python
from PIL import Image
from matplotlib import pyplot as plt
import random
from pathlib import Path
import tensorflow as tf
import numpy as np
```

```python
img_path = '/Volumes/T5/Datasets/FFHQ/69918.png'
```

```python

im = Image.open(img_path)
im = np.array(im)
plt.imshow(im)
```

```python
tensor = np.array(im)
# tensor = tf.convert_to_tensor(tensor)
tensor = tf.keras.preprocessing.image.random_rotation(tensor, 25, fill_mode='reflect', row_axis=0, col_axis=1, channel_axis=2)
zoom_ratio = random.random() * 1 + 0.75 
tensor = tf.keras.preprocessing.image.random_zoom(tensor, (zoom_ratio, zoom_ratio), fill_mode='reflect', row_axis=0, col_axis=1, channel_axis=2)
tensor = np.array(tensor)

plt.imshow(tensor)
```
