```python
from PIL import Image
from matplotlib import pyplot as plt
import random
from pathlib import Path
```

```python
img_path = '/Volumes/T5/Workspace/bosch/The_Garden_of_Earthly_Delights_by_Bosch_High_Resolution.jpg'
out_dir = Path('../datasets/bosch')
out_dir.mkdir(exist_ok=True)
```

```python

Image.MAX_IMAGE_PIXELS = 1000000000
im = Image.open('/Volumes/T5/Workspace/bosch/The_Garden_of_Earthly_Delights_by_Bosch_High_Resolution.jpg')

scale = 0.5
im = im.resize((int(im.width * scale), int(im.height * scale)), Image.LANCZOS)
print(im.size)
```

```python
def random_crop(im, width, height):
    x = random.randrange(im.width - width)
    y = random.randrange(im.height - height)
    cropped = im.crop((x, y, x+width, y+height))
    return cropped

plt.imshow(random_crop(im, 512, 512))
```

```python
for i in range(500):
    cropped = random_crop(im, 512, 512)
    cropped.save(out_dir / f'{i:03d}.png')
```

```python
for i in range(500, 1000):
    cropped = random_crop(im, 1024, 1024)
    cropped = cropped.resize((512, 512), Image.LANCZOS)
    cropped.save(out_dir / f'{i:03d}.png')
```
